```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
fp = r"C:\Users\A\Box\EEP147\2024\ESG\Stata\Bids\2024 EEP 147 ESG Permit Auctions (Responses) - Round4.csv"
df = pd.read_csv(fp)
df
```

```python
supply = df[df.buy_sell == "sell"]
demand = df[df.buy_sell == "buy"]
```

```python
# Plot both on same graph
x = 'permits'
y = 'price'
plt.scatter(supply[x], supply[y], color='red', label='supply')
plt.scatter(demand[x], demand[y], color='blue', label='demand')
plt.xlabel('permits')
plt.ylabel('price')
plt.legend()
plt.show()
```
